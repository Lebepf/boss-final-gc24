S03T03 - Transformar variáveis categóricas em numéricas (one-hot encoding, label encoding, etc.).

```python
import pandas as pd

df_heart = pd.read_csv('heart_attack_prediction_dataset.csv')
df_heart.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Identificando as colunas categóricas
colunas_label_encoding = ['Sex', 'Hemisphere']  # Poucas categorias
colunas_one_hot_encoding = ['Country', 'Continent']  # Muitas categorias

# 1. Aplicando Label Encoding para colunas com poucas categorias
labelencoder = LabelEncoder()
for coluna in colunas_label_encoding:
    df_heart[coluna] = labelencoder.fit_transform(df_heart[coluna])

# 2. Aplicando One-Hot Encoding para colunas com muitas categorias
df_heart = pd.get_dummies(df_heart, columns=colunas_one_hot_encoding, drop_first=True)

# Verificando o resultado final
print(df_heart.info())
print(df_heart.head())
```

```python
# Ajustar para exibir mais colunas
pd.set_option('display.max_columns', None)  # Mostra todas as colunas
pd.set_option('display.width', 1000)       # Ajusta a largura da saída
print(df_heart.head())
```

```python
# Exibir as primeiras 10 colunas
print(df_heart.iloc[:, :10].head())

# Exibir da 11ª até a 20ª coluna
print(df_heart.iloc[:, 10:20].head())
```
