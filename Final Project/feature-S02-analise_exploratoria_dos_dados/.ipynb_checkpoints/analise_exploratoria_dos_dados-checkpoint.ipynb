# Identificar e Tratar Outliers

## Calcular o Intervalo Interquartil (IQR) para cada coluna numérica

```python
Q1 = data[numerical_columns].quantile(0.25)  # Primeiro quartil (25%)
Q3 = data[numerical_columns].quantile(0.75)  # Terceiro quartil (75%)
IQR = Q3 - Q1
```

## Limites para identificar outliers

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

## Identificar outliers

```python
outliers = ((data[numerical_columns] < lower_bound) | (data[numerical_columns] > upper_bound))
print("\nNúmero de Outliers por Coluna:")
print(outliers.sum())
```

## Tratar outliers

```python
data_no_outliers = data[~outliers.any(axis=1)]
```

## Exibir a redução do dataset

```python
print(f"\nDimensão original do dataset: {data.shape}")
print(f"Dimensão após remoção de outliers: {data_no_outliers.shape}")
```

## Identificar colunas numéricas

# Visualizações:

## Histograma

```python
for column in numerical_columns[:3]:
    plt.figure(figsize=(8, 5))
    data[column].hist(bins=30, edgecolor='black')
    plt.title(f'Histograma de {column}')
    plt.xlabel(column)
    plt.ylabel('Frequência')
    plt.show()
```

## Boxplot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[numerical_columns])
plt.title('Boxplots de Colunas Numéricas')
plt.xticks(rotation=45)
plt.show()
```

## Mapa de calor de correlação

```python
plt.figure(figsize=(10, 8))
correlation = data[numerical_columns].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor das Correlações')
plt.show()
```

# Arquivos:

## Salvar o DataFrame limpo

```python
data.to_csv('data/dados_limpos.csv', index=False)
print("\nDados limpos salvos como 'dados_limpos.csv'.")
```

## Gerar e salvar o relatório como PDF

```python
cleaned_data = data

# Criar o PDF para o relatório
pdf_path = 'data/EDA_Cleaned_Heart_Disease_Report.pdf'
with PdfPages(pdf_path) as pdf:
    
    # Página 1: Estatísticas Descritivas
    plt.figure(figsize=(10, 6))
    plt.text(0.5, 0.9, "Exploração de Dados (EDA): Relatório Inicial", fontsize=20, ha='center')
    plt.text(0.05, 0.75, "Insights Baseados nas Estatísticas Descritivas:", fontsize=14, va='top')
    insights = """
    1. A média de idade é cerca de 54 anos, abrangendo 29 a 77 anos.
    2. A pressão arterial média em repouso é 131 mmHg; valores sugerem casos de hipertensão.
    3. O colesterol médio é 247 mg/dL, com valores elevados indicando alto risco cardiovascular.
    4. Frequência cardíaca máxima média de 149 bpm, reduzindo com a idade (-0.39 correlação).
    5. Cerca de 14% dos pacientes têm níveis elevados de açúcar no sangue.
    """
    plt.text(0.05, 0.05, insights, fontsize=12, va='top')
    plt.axis('off')
    pdf.savefig()
    plt.close()

    # Página 2: Histogramas
    for column in ["age", "trtbps", "chol", "thalachh"]:
        plt.figure(figsize=(8, 5))
        sns.histplot(cleaned_data[column], kde=True, color='blue', bins=20)
        plt.title(f"Distribuição de {column}", fontsize=16)
        plt.xlabel(column)
        plt.ylabel("Frequência")
        pdf.savefig()
        plt.close()

    # Página 3: Boxplots
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=cleaned_data[["age", "trtbps", "chol", "thalachh"]])
    plt.title("Boxplot das Variáveis Contínuas", fontsize=16)
    plt.xticks(rotation=45)
    pdf.savefig()
    plt.close()

    # Página 4: Correlação
    plt.figure(figsize=(10, 8))
    sns.heatmap(cleaned_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title("Mapa de Correlação", fontsize=16)
    pdf.savefig()
    plt.close()

    # Página 5: Conclusões
    plt.figure(figsize=(10, 6))
    plt.text(0.05, 0.9, "Conclusões e Próximos Passos:", fontsize=14, va='top')
    conclusions = """
    1. Investigação adicional é necessária para padrões sazonais ou de idade em doenças cardíacas.
    2. Outliers em colesterol e pressão arterial foram tratados no dataset limpo.
    3. Analisar detalhadamente os grupos de risco usando variáveis categóricas como 'cp'.
    """
    plt.text(0.05, 0.7, conclusions, fontsize=12, va='top')
    plt.axis('off')
    pdf.savefig()
    plt.close()

print(f"Relatório salvo em: {pdf_path}")
```
