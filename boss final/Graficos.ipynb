```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
# Carregar os dados
df_heart = pd.read_csv('heart_attack_prediction_dataset.csv')
df_heart.head()
```

```python
# Definindo o estilo dos gráficos
sns.set(style="whitegrid")

# Gráfico 1: Distribuição de variáveis numéricas
plt.figure(figsize=(10, 6))
sns.histplot(df_heart['BMI'], kde=True, color='blue', bins=30)
plt.title('Distribuição do BMI')
plt.xlabel('BMI')
plt.ylabel('Frequência')
plt.show()
```

```python
# Boxplot de pressão arterial
plt.figure(figsize=(10, 6))
sns.boxplot(x='Blood Pressure', data=df_heart)
plt.title('Boxplot de Pressão Arterial')
plt.xlabel('Pressão Arterial')
plt.show()
```

```python
#Distribuição de variáveis categóricas
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df_heart, hue='Sex', dodge=False)
plt.title('Distribuição de Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.legend([], [], frameon=False)  # Remove legenda redundante
plt.show()
```

```python

# Dados de exemplo para a matriz de acurácia
accuracy_matrix = pd.DataFrame({
    "Dataset": ["Treinamento", "Teste"],
    "Acurácia": [100, 0.81] 
})

# Configurar o índice
accuracy_matrix.set_index("Dataset", inplace=True)

# Gráfico de barras
plt.figure(figsize=(8, 6))
accuracy_matrix.plot(kind='bar', color=['lightblue', 'lightgreen'], legend=False)
plt.title('Comparação de Acurácia: Treinamento vs Teste')
plt.ylabel('Acurácia')
plt.ylim(0, 1) 
plt.show()

```

```python

Treinamento = [100] 
Teste = [81] 

# Gerar a matriz de confusão
conf_matrix = confusion_matrix(Treinamento, Teste)

# Criar o heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues', cbar=False,
            xticklabels=['Predito: Não Saudável', 'Predito: Saudável'],
            yticklabels=['Real: Não Saudável', 'Real: Saudável'])
plt.title('Matriz de Confusão')
plt.xlabel('Predição')
plt.ylabel('Valor Real')
plt.show()
```
