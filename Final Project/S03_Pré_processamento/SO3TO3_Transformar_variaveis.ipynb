### S03T03 - Transformar variáveis categóricas em numéricas (one-hot encoding, label encoding, etc.).

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Carregar o dataset
df_heart = pd.read_csv("heart_attack_prediction_dataset.csv")
df_heart.head()
```

```python
# Verificando os nomes das colunas
print(df_heart.columns)  # Exibe todas as colunas do DataFrame
```

```python
# Identificando as colunas categóricas restantes
colunas_label_encoding = ['Sex', 'Hemisphere']  # Colunas com poucas categorias
colunas_label_encoding
```

```python
# 1. Aplicando Label Encoding para colunas com poucas categorias
labelencoder = LabelEncoder()
for coluna in colunas_label_encoding:
    df_heart[coluna] = labelencoder.fit_transform(df_heart[coluna])
df_heart.head()
```

```python
#aplicando o One-Hot Encoding em outras colunas categóricas ainda não tratadas
colunas_one_hot_encoding = ['Diet', 'Smoking', 'Previous Heart Problems', 'Medication Use', 'Stress Level']  
colunas_one_hot_encoding
```

```python
df_heart = pd.get_dummies(df_heart, columns=colunas_one_hot_encoding, drop_first=True)
df_heart
```

```python
df_heart.columns
```

```python
# Atualize a lista de colunas a serem codificadas, excluindo as colunas já codificadas
colunas_one_hot_encoding = [
    'Diet_Unhealthy',  # Caso você tenha uma coluna 'Diet' que não tenha sido codificada
    'Smoking_1',  # Se houver uma coluna 'Smoking' sem o sufixo '_1'
    'Previous Heart Problems_1',  # Se não houver uma coluna codificada para isso
    'Medication Use_1',  # Caso não tenha sido codificada ainda
    'Stress Level_10'  # Se houver uma coluna sem o sufixo '_2', '_3', etc.
]

# Aplicar o one-hot encoding nas colunas ajustadas
df_heart = pd.get_dummies(df_heart, columns=colunas_one_hot_encoding, drop_first=True)

# Verifique o resultado
print(df_heart.columns)
```

```python
# Verificando o resultado final
print(df_heart.info())  # Informações sobre o DataFrame
print(df_heart.head())  # Primeiras 5 linhas após a transformação
```
